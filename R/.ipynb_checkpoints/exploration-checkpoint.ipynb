```r
data = read_csv("data.csv")
str(data)
```

```r
head(data)
```

The dataset is composed of 5 features and one binary target.

Each transaction has an unique identifier, the date of transaction, the costumer unique id, a termianl id that identify the 'location' where the transaction is done and the amount of the transaction.

Order of exploration:

1) full dataset \
2) costumer profile \
3) terminal profile \

```r
# plot the total number of transaction by day

tx_byday = data %>%
            mutate(day = date(TX_DATETIME)) %>%
            group_by(day) %>%
            summarise(
                n_transactions = n(),
                n_fraud = sum(TX_FRAUD),
                mean_amount = mean(TX_AMOUNT)
            )
head(tx_byday)
```

```r
ggplot(data = tx_byday) +
    geom_line(mapping = aes(x=day, y=n_fraud, color="# fraud transaction by day")) +
    geom_line(mapping = aes(x=day, y=n_transactions/50, color="# transaction by day / 50")) +
    labs(x="Date", y="Number of transaction") +
    scale_x_date(date_breaks = "1 month") 
    
```

```r
summary(tx_byday)
```

we have a preiod with an increasing in number of fraudulent transaction and then it stabilize at 80 fraud per day. The nuber of transaction folllow a stationary process with mean of 9500 transactions per day.

```r
vc = data %>% group_by(TX_FRAUD) %>% summarise(count = n() / dim(data)[1])
ggplot(data=vc) +
    geom_col(mapping=aes(x=TX_FRAUD, y=count))
vc
```

The datset present only 0.008% of fraudulent transaction.

It must be take into account this fact in the modelling part and in the evaluation metric.

```r
# let's explore the tx amount
library(scales)

summary(data$TX_AMOUNT)
ggplot(data=data[sample(nrow(data),10000),], mapping=aes(x=TX_AMOUNT)) + geom_histogram(binwidth=10)
ggplot(data=data, mapping=aes(y=TX_AMOUNT)) + geom_boxplot() 
```

the majoprity of transaction are with a small amount, as the amount increase the number of transaction decrese very rapidly. transaction with an amount >=250 are veryr are.

```r
ggplot(data=data, mapping=aes(y=TX_AMOUNT, fill=as.factor(TX_FRAUD))) + geom_boxplot() 
```

from the boxplot is clear that transaction with a very high amount are more likely to be fraud w.r.t transaction with a small amount. this represent the fact that when a card is "stolen" the thief want to amximize thei profit by retrive as much as possible fro the stolen credit card

```r
# plot distribution of transaction based on day and time to see what 
# happen during week end and in if there are pattern based on time

sample_data = data # [sample(nrow(data),10000),]

tx_bydate = sample_data %>%
    mutate(day = wday(TX_DATETIME), hour = hour(TX_DATETIME)) %>%
    group_by(day, hour) %>%
    summarise(
                n_transactions = n(),
                n_fraud = sum(TX_FRAUD),
                mean_amount = mean(TX_AMOUNT)
            )
head(tx_bydate)
```

```r
ggplot(data=tx_bydate) +
    geom_col(mapping=aes(x=hour, y=n_transactions))
ggplot(data=tx_bydate) +
    geom_col(mapping=aes(x=hour, y=n_fraud))
```

```r
ggplot(data=tx_bydate) +
    geom_col(mapping=aes(x=day, y=n_transactions))
ggplot(data=tx_bydate) +
    geom_col(mapping=aes(x=day, y=n_fraud))
```
