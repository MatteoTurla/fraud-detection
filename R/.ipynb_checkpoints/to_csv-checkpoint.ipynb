```python
import pickle as pk
import pandas as pd
import numpy as np
import os
```

```python
directory = "simulated-data-raw/data/"
list_df = []
for filename in os.listdir("simulated-data-raw/data/"):
    file = directory+filename
    list_df.append(pd.read_pickle(file))
df = pd.concat(list_df, axis=0)
df = df.set_index("TRANSACTION_ID").reset_index()
```

```python
df.head()
```

```python
# we need to remove simulated column
drop = ["TX_TIME_SECONDS", "TX_TIME_DAYS", "TX_FRAUD_SCENARIO"]
df = df.drop(drop, axis=1)
df.head()
```

```python
df.info()
```

```python
df.to_csv("data.csv", index=False)
```
